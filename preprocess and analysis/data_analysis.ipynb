```python
data_df1=pd.read_csv("./CSSEcovid19/time_series_covid19_confirmed_global.csv")
data_df2=pd.read_csv("./CSSEcovid19/time_series_covid19_deaths_global.csv")

dates=data_df1.columns[4:]

hubei1=data_df1.loc[72]
hubei2=data_df2.loc[72]

print(hubei1["Province/State"])

Y1=[hubei1[dates[0]]]
Y2=[hubei2[dates[0]]]
for day in range(1,len(dates)):
    Y1.append(hubei1[dates[day]]-hubei1[dates[day-1]])
    Y2.append(hubei2[dates[day]]-hubei2[dates[day-1]])

drawPlot(pd.DataFrame(Y1,index=range(1,len(dates)+1),columns=['value']),title="Hubei confirmed cases")
drawPlot(pd.DataFrame(Y2,index=range(1,len(dates)+1),columns=['value']),title="Hubei deaths")
```
